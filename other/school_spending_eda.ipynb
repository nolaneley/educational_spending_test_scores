### Fit linear regression to spendingmath relationship

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.feature_selection import f_regression
from sklearn.feature_selection import r_regression

# Convert to numpy arrays for easier processing
spending = np.array(df.spending)
math = np.array(df.math)

# Expand dimensions to fit into sklearn regression
spending = np.expand_dims(spending, axis=1)
math = np.expand_dims(math, axis=1)

# Create regression object
regression = LinearRegression()
regression.fit(spending, math)

# Plot scatter plot with regression line
x = np.linspace(min(spending),max(spending),50)
plt.scatter(spending, math, color='#5F7DA5')
plt.plot(x, regression.predict(x), color='#B07571')
plt.title("Spending versus outcome in MATH")
plt.xlabel("Median per pupil spending ($)")
plt.ylabel("Average math score")

# Calculate and print MSE and slope
mse = mean_squared_error(math, regression.predict(spending))
print('Mean Squared Error: ', mse)
print('Slope: ', regression.coef_[0])

# Calculate f-stat and p-value

r_math = r_regression(spending,math)
_, p_math = f_regression(spending,math)

print("r math: ", str(r_math))
print("p math: ", str(p_math))

```

### Fit linear regression to spendingreading relationship

```python
# Get reading scores as array and reshape
reading = np.array(df.reading)
reading = np.expand_dims(reading, axis=1)

# Create regression object
regression = LinearRegression()
regression.fit(spending, reading)

# Plot scatter plot with regression line
plt.scatter(spending, reading, color='#5F7DA5')
plt.plot(spending, regression.predict(spending), color='#B07571')
plt.title("Spending versus outcome in READING")
plt.xlabel("Median per pupil spending ($)")
plt.ylabel("Average reading score")

# Calculate and print MSE and slope
mse = mean_squared_error(reading, regression.predict(spending))
print('Mean Squared Error: ', mse)
print('Slope: ', regression.coef_[0])

# Calculate f stat and p-value
r_reading = r_regression(spending, reading)
_, p_reading = f_regression(spending,reading)
print("r reading: ", str(r_reading))
print("p reading: ", str(p_reading))
```

### Find 10 states with best ratio of spending to math and reading outcomes

```python
# Calculate ratios and get sorted indices to apply to states list
rat_math = np.squeeze(spending) / np.squeeze(math)
rat_reading = np.squeeze(spending) / np.squeeze(reading)
sorted_indices_math = sorted(range(len(rat_math)), key=lambda i: rat_math[i])
sorted_indices_read = sorted(range(len(rat_reading)), key=lambda i: rat_reading[i])

states = df.state # list of states in original (arbitrary)

# Show top ratios for math and reading
states_math_sorted = [states[i] for i in sorted_indices_math]
print("\n10 best spending-to-outcome ratio in math:")
[print(state) for state in states_math_sorted[0:10]]

states_read_sorted = [states[i] for i in sorted_indices_read]
print("\n10 best spending-to-outcome ratio in reading:")
[print(state) for state in states_read_sorted[0:10]]
```

### Find 10 states with worst ratio of spending to math and reading outcomes

```python
# Show top ratios for math and reading
states_math_sorted = [states[i] for i in sorted_indices_math]
print("\n10 worst spending-to-outcome ratio in math:")
[print(state) for state in states_math_sorted[-10:][::-1]]

states_read_sorted = [states[i] for i in sorted_indices_read]
print("\n10 worst spending-to-outcome ratio in reading:")
[print(state) for state in states_read_sorted[-10:][::-1]]
```
